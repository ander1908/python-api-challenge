### Humidity Heatmap

* Use the Lat and Lng as locations and Humidity as the weight.

```python
#ABOVE WAS THE TEST FOR THE ITERATION BELOW

hotel_name = []
types = "lodging"
radius = 5000
for index,row in hotel_df.iterrows():

    lat = row["Latitude"]
    lon = row["Longitude"]
    city = row["Cities"]
    
    params = {"location": f"{lat},{lon}", "types": types,"radius": radius,"key": g_key,}
    base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
    
    response = requests.get(base_url,params=params)
    results = response.json()
    #print(json.dumps(results, indent=4,sort_keys=True))
    try:
        print(f"Working on finding a hotel in {city}")
        hotel_df.loc[index,"Hotel Name"] = results["results"][0]["name"]
    except (KeyError, IndexError):
        print("Missing field/result... skipping.")
        hotel_df.loc[index, "Hotel Name"] = "N/A"
     
```
